```python
import pandas as pd

url = "https://raw.githubusercontent.com/asharvi1/UCI-Air-Quality-Data/master/AirQualityUCI.csv"
df = pd.read_csv(url)

df = pd.read_csv(url, sep=';')

# Drop empty columns
df = df.dropna(axis=1, how='all')

# Replace -200 with NaN
df = df.replace(-200, pd.NA)

# Drop missing rows
df = df.dropna()

df.head()

import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip().str.lower()

# Fix time format (VERY IMPORTANT)
df['time'] = df['time'].str.replace('.', ':', regex=False)

# Create datetime
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], dayfirst=True)

df.head()

import matplotlib.pyplot as plt

df = df.sort_values('datetime')

plt.figure(figsize=(10,5))
plt.plot(df['datetime'], df['no2(gt)'])
plt.title("NO2 Levels Over Time")
plt.xlabel("Date")
plt.ylabel("NO2")
plt.show()

plt.figure()
plt.scatter(df['t'], df['no2(gt)'])
plt.xlabel("Temperature")
plt.ylabel("NO2")
plt.title("Temperature vs NO2")
plt.show()
```

## 🔍 Exploratory Data Analysis

- NO2 levels vary significantly over time, indicating fluctuations in air pollution  
- There is a slight relationship between temperature and NO2 levels, though not strongly linear  
- Pollution patterns suggest environmental and temporal influences
